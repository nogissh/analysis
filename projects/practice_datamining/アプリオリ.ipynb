```python
import pandas as pd

df = pd.read_csv('./datasets/sales_reciepts.csv')
df.head()
```

```python
dummy = pd.get_dummies(df['product_id'], prefix='prd_')
ddf = pd.DataFrame(df['transaction_id'])
ddf[dummy.columns] = dummy.values
```

```python
import numpy as np

r = []
for t_id in df['transaction_id'].unique():
    r.append((np.sum(ddf[ddf['transaction_id'] == t_id].drop(['transaction_id'], axis=1), axis=0).values > 0).astype('int8'))
```

```python
dddf = pd.DataFrame(r, columns=dummy.columns)
dddf['transaction_id'] = df['transaction_id'].unique().astype('int8')
```

```python
from mlxtend.frequent_patterns import apriori

r_apr = apriori(dddf.drop('transaction_id', axis=1), min_support=0.03, use_colnames=True)
```

```python
r_apr[r_apr['itemsets'] & frozenset(['prd__87'])].sort_values('support', ascending=False)
```
