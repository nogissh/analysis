## ロジスティック回帰

```python
import pandas as pd

loan_data_df = pd.read_csv('./datasets/loan_data.csv')
loan_data_df.head()
```

```python
predictors = ['payment_inc_ratio', 'purpose_', 'home_', 'emp_len_', 'borrower_score']
outcome = 'outcome'
```

```python
X = pd.get_dummies(loan_data_df[predictors], prefix='', prefix_sep='', drop_first=True)
y = loan_data_df[outcome]
```

```python
from sklearn.linear_model import LogisticRegression

logit_reg = LogisticRegression(penalty='l2', C=1e42, solver='liblinear')
logit_reg.fit(X, y)
```

```python
# 分布の特徴を取得する
pred = pd.DataFrame(logit_reg.predict_log_proba(X), columns=loan_data_df[outcome].astype('category').cat.categories)
pred.describe()
```

```python
# 確率を取得する
pred_ = pd.DataFrame(logit_reg.predict_proba(X), columns=loan_data_df[outcome].astype('category').cat.categories)
pred_.describe()
```
