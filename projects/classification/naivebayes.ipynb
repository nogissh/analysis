## ナイーブベイズ

```python
import pandas as pd

df = pd.read_csv('./datasets/loan_data.csv')
df.head()
```

```python
from sklearn.naive_bayes import MultinomialNB

predictors = ['purpose_', 'home_', 'emp_len_']
outcome = 'outcome'

X = pd.get_dummies(df[predictors], prefix='', prefix_sep='')
y = df[outcome]

naive_model = MultinomialNB(alpha=0.01, fit_prior=True)
naive_model.fit(X, y)
```

```python
new_loan = X.loc[146:146, :]
new_loan.head()
```

```python
print('predicted class', naive_model.predict(new_loan)[0])

# テキスト記載のままでは動かない
# 参照 https://github.com/rajewsky-lab/planarian_lineages/issues/2
probabilities = pd.DataFrame(naive_model.predict_proba(new_loan), columns=df[outcome].astype('category').cat.categories)
probabilities
```
